# U.S. Medical Insurance Costs

```python
import pandas as pd
```

```python
dataset = pd.read_csv('insurance.csv')
dataset
```

```python
dataset.columns
```

```python
#counting if there are any null values within the different fields
age_nan_count = dataset['age'].isna().sum()
sex_nan_count = dataset['sex'].isna().sum()
bmi_nan_count = dataset['bmi'].isna().sum()
children_nan_count = dataset['children'].isna().sum()
smoker_nan_count = dataset['smoker'].isna().sum()
region_nan_count = dataset['region'].isna().sum()
charges_nan_count = dataset['charges'].isna().sum()

print("age =", age_nan_count)
print("sex =", sex_nan_count)
print("bmi =", bmi_nan_count)
print("children =", children_nan_count)
print("smoker =", smoker_nan_count)
print("region =", region_nan_count)
print("charges =", charges_nan_count)
```

```python
#identifying the different regions
dataset.region.unique()
```

```python
#mean age in the dataset
mean_age = dataset['age'].mean();
print(mean_age)
```

```python
#splitting the dataset into the different regions
southwest = dataset[dataset['region']=='southwest']
southeast = dataset[dataset['region']=='southeast']
northwest = dataset[dataset['region']=='northwest']
northeast = dataset[dataset['region']=='northeast']
```

```python
#counting the number of respondents in each region
southwest_count = len(southwest);
southeast_count = len(southeast);
northwest_count = len(northwest);
northeast_count = len(northeast);

print(southwest_count)
print(southeast_count)
print(northwest_count)
print(northeast_count)
```

```python
#average bmi in each region
sw_mean_bmi = southwest['bmi'].mean();
se_mean_bmi = southeast['bmi'].mean();
nw_mean_bmi = northwest['bmi'].mean();
ne_mean_bmi = northeast['bmi'].mean();

print(sw_mean_bmi)
print(se_mean_bmi)
print(nw_mean_bmi)
print(ne_mean_bmi)
```

```python
#average charge per region
sw_mean_charge = southwest['charges'].mean();
se_mean_charge = southeast['charges'].mean();
nw_mean_charge = northwest['charges'].mean();
ne_mean_charge = northeast['charges'].mean();

print(sw_mean_charge)
print(se_mean_charge)
print(nw_mean_charge)
print(ne_mean_charge)
```

```python
#average number of children in each region
sw_mean_children = southwest['children'].mean();
se_mean_children = southeast['children'].mean();
nw_mean_children = northwest['children'].mean();
ne_mean_children = northeast['children'].mean();

print(sw_mean_children)
print(se_mean_children)
print(nw_mean_children)
print(ne_mean_children)
```

```python
#average age in each region
sw_mean_age = southwest['age'].mean();
se_mean_age = southeast['age'].mean();
nw_mean_age = northwest['age'].mean();
ne_mean_age = northeast['age'].mean();

print(sw_mean_age)
print(se_mean_age)
print(nw_mean_age)
print(ne_mean_age)
```

```python
#histogram bmi based on if they are smokers or not
smoker = dataset[dataset['smoker']=='yes'];
non_smoker = dataset[dataset['smoker']=='no'];

sns.set_theme(style="darkgrid")
sns.histplot(smoker['bmi'], kde=True).set(title='Histogram on respondents who smoke');
```

```python
sns.set_theme(style="darkgrid")
sns.histplot(non_smoker['bmi'], kde=True).set(title='Histogram on respondents who do not smoke');
```

```python
smoker_mean_bmi = smoker['bmi'].mean();
non_smoker_mean_bmi = non_smoker['bmi'].mean();

print(smoker_mean_bmi)
print(non_smoker_mean_bmi)
```

```python
#mean bmi for males and females
male = dataset[dataset['sex']=='male'];
female = dataset[dataset['sex']=='female'];

male_mean_bmi= male['bmi'].mean();
female_mean_bmi= female['bmi'].mean();

print(male_mean_bmi)
print(female_mean_bmi)
```
